```python
import networkx as nx
```

```python
map = "XXXXXXXXXXXX\nX...X..X..eX\nX.X....X.XXX\nX..X.X.X.X.X\nXX.XXX.X...X\nX........X.X\nXsXX...X...X\nXXXXXXXXXXXX"

print(map)
```

```python
map_split = map.split("\n")
print(map_split)
```

```python
# Create an empty graph
maze_graph = nx.Graph()
```

```python
# Add the nodes (cells) to the graph
for i in range(len(map_split)):
    for j in range(len(map_split[0])):
        if map_split[i][j] != 'X': #1 is used to represent walls
            maze_graph.add_node((i, j))
```

```python
nx.draw(maze_graph, pos=nx.spring_layout(maze_graph), with_labels=True, node_size=500)
```

```python
# Add the edges (connections) to the graph
for node in maze_graph.nodes:
    row, col = node

    # Check the north cell
    if row > 0 and map_split[row-1][col] != 'X':
        maze_graph.add_edge(node, (row-1, col))

    # Check the west cell
    if col > 0 and map_split[row][col-1] != 'X':
        maze_graph.add_edge(node, (row, col-1))

    # Check the south cell
    if row < len(map_split)-1 and map_split[row+1][col] != 'X':
        maze_graph.add_edge(node, (row+1, col))

    # Check the east cell
    if col < len(map_split[0])-1 and map_split[row][col+1] != 'X':
        maze_graph.add_edge(node, (row, col+1))
```

```python
nx.draw(maze_graph, pos=nx.spring_layout(maze_graph), with_labels=True, node_size=500)
```
